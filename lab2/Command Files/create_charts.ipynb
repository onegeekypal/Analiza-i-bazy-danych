**imports and data**

```python
import pandas
import matplotlib.pyplot as plt

age_gender_worried_table = pandas.read_csv('/Users/dawidsyty/DataspellProjects/Analiza-i-bazy-danych/lab2/Analysis Data/age_gender_worried_table.csv')
renamed_question_table = pandas.read_csv('/Users/dawidsyty/DataspellProjects/Analiza-i-bazy-danych/lab2/Analysis Data/question_renamed.csv')
```

**Answers chars**

```python
for column in renamed_question_table:
    answers = renamed_question_table[column].value_counts().to_dict()
    plt.figure(figsize=(20,10))
    plt.title(column)
    plt.bar(answers.keys(),answers.values())
    plt.savefig('/Users/dawidsyty/DataspellProjects/Analiza-i-bazy-danych/lab2/Analysis Data/charts/' + column + ".png")
    plt.show()
```
